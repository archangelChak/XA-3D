```python
coeff = 2j*np.pi/S_swap.shape[1]
coeff_xy = np.pi/S_swap.shape[0]
sin_m = np.sin(np.arange(S_swap.shape[0])*coeff_xy)
cos_m = np.cos(np.arange(S_swap.shape[0])*coeff_xy)
```

```python
exp = np.exp(exp_power)
```

```python
res = np.empty((frames_remote.shape[1],100,100))
```

```python
for z in tqdm(range(frames_remote.shape[1])):
    frames_remote_zero = frames_remote[:,z]
    S_swap = np.empty_like(frames_remote_zero, dtype=np.complex64)
    for i in range(num_frames_remote):
        S_swap[i] = np.fft.fft(frames_remote_zero[i], norm = "ortho")
    for i in range(S_swap.shape[-1]):
        S_swap[:,i] *= i/S_swap.shape[-1]
    res_nosum = np.empty((100,100,122,960),dtype = np.complex64)
    for x in range(100):
        for y in range(100):
            res_nosum[x,y] = exp[x,y]*S_swap
    res[z] = res_nosum.sum(axis=(2,3))
```
