```python
frames_remote_swap = np.swapaxes(frames_remote,0,1)
frames_remote_swap.shape
```

```python
S_swap = np.empty_like(frames_remote_swap, dtype=np.complex128)
for i in range(frames_remote.shape[0]):
    for j in range(num_frames_remote):
        S_swap[:][i][j] = np.fft.fft(frames_remote[i][j])
for i in range(S_swap.shape[-1]):
    S_swap[:,:,i] *= i/S_swap.shape[-1]
```

```python
coeff = 2j*np.pi/S_swap.shape[2]
coeff_xy = np.pi/S_swap.shape[1]
sin_m = np.sin(np.arange(S_swap.shape[0])*coeff_xy)
cos_m = np.cos(np.arange(S_swap.shape[0])*coeff_xy)
```

```python
exp_power = np.empty((50,50,S_swap.shape[1],S_swap.shape[2]),dtype = np.complex64)
for n in tqdm(range(-480,480)):
    for x in range(50):
        for y in range(50):
            for m in range(S_swap.shape[1]):
                exp_power[x,y,m,n+480] = n*coeff*(x*cos_m[m]+y*sin_m[m])
```
