```python
import numpy as np
import os
import pydicom
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
dicom_dir = "C:/123/XA-3D"
dicom_files = [os.path.join(dicom_dir, f) for f in os.listdir(dicom_dir) if (f.startswith('im') or f.endswith(".dcm"))]
```

```python
ds_example =  pydicom.dcmread(dicom_files[0])
ds_ct = pydicom.dcmread(dicom_files[1])
ds_remote = pydicom.dcmread(dicom_files[2])
```

```python
frames_remote = ds_remote.pixel_array
num_frames_remote = ds_remote.NumberOfFrames
frames_сt = ds_ct.pixel_array
num_frames_ct = ds_ct.NumberOfFrames
```

```python
frames_remote_zero = frames_remote[:,370]
```

```python
S_swap = np.empty_like(frames_remote_zero, dtype=np.complex64)
for i in range(num_frames_remote):
    S_swap[i] = np.fft.fft(frames_remote_zero[i])
for i in range(S_swap.shape[-1]):
    S_swap[:,i] *= i/S_swap.shape[-1]
```

```python
coeff = 2j*np.pi/S_swap.shape[1]
coeff_xy = np.pi/S_swap.shape[0]
sin_m = np.sin(np.arange(S_swap.shape[0])*coeff_xy)
cos_m = np.cos(np.arange(S_swap.shape[0])*coeff_xy)
```

```python
f = np.zeros((frames_remote_zero.shape[1],frames_remote_zero.shape[1]))
for m in tqdm(range(num_frames_remote)):
    for x in range(frames_remote_zero.shape[1]):
        for y in range(frames_remote_zero.shape[1]):
            s = x*sin_m[m] - y/cos_m[m]
            if (abs(s) < 960):
                q = S_swap[m,int(abs(s))] * np.exp(2j*np.pi*s*(x*cos_m[m]+y*sin_m[m]))
                f[x,y] += q
```
